```python
import cv2
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.image import MIMEImage
from ultralytics import YOLO

# Load the trained model
model = YOLO(r"C:\Users\Administrator\Downloads\best fire.pt")

# Open video file or webcam
cap = cv2.VideoCapture(r"C:\Users\Administrator\Downloads\fires video.mp4")  

# Email settings
sender_email = "janaelpardisi@gmail.com"
receiver_email = "receiver_email@gmail.com"
password = "olpn fpwp tcky dpsm"  # Use App Password (not your real password)

def send_email(image):
    msg = MIMEMultipart()
    msg['From'] = sender_email
    msg['To'] = receiver_email
    msg['Subject'] = 'Fire & Smoke Detection Image'

    # Add text message
    body = MIMEText("Detected fire or smoke in the image.")
    msg.attach(body)

    # Attach image
    image_data = cv2.imencode('.jpg', image)[1].tobytes()
    image_mime = MIMEImage(image_data, name="detection.jpg")
    msg.attach(image_mime)

    try:
        server = smtplib.SMTP('smtp.gmail.com', 587)
        server.starttls()
        server.login(sender_email, password)
        server.sendmail(sender_email, receiver_email, msg.as_string())
        server.quit()
        print("Image sent via email.")
    except Exception as e:
        print(f"Failed to send email: {e}")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Resize the frame
    frame = cv2.resize(frame, (1020, 500))

    # Run detection
    results = model(frame)

    # Draw detection results on the frame
    for r in results:
        boxes = r.boxes
        for box in boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = box.conf[0]
            cls = int(box.cls[0])
            label = model.names[cls]

            # Color based on label
            color = (0, 0, 255) if label == "fire" else (255, 255, 0)

            # Shrink the bounding box to improve visibility
            margin = 10
            x1, y1, x2, y2 = x1 + margin, y1 + margin, x2 - margin, y2 - margin

            # Draw a thicker bounding box
            box_thickness = 4
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, box_thickness)

            # Prepare label text
            font = cv2.FONT_HERSHEY_SIMPLEX
            font_scale = 0.9
            thickness = 2
            text = f"{label} {conf:.2f}"
            text_size = cv2.getTextSize(text, font, font_scale, thickness)[0]
            text_w, text_h = text_size

            # Position the text above or below the box
            text_x = x1
            text_y = y1 - 10 if y1 - 10 > 10 else y1 + text_h + 10

            # Draw background for the text
            cv2.rectangle(frame, (text_x, text_y - text_h - 5), (text_x + text_w, text_y), color, -1)
            cv2.putText(frame, text, (text_x, text_y), font, font_scale, (255, 255, 255), thickness)

    # Show the detection window
    cv2.imshow("Fire & Smoke Detection", frame)

    # Send detection image by email
    send_email(frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
